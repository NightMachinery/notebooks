# Model

- [GRU --- PyTorch 2.0 documentation](https://pytorch.org/docs/stable/generated/torch.nn.GRU.html)

- [torch.nn.utils.rnn.pack~sequence~ --- PyTorch 2.0 documentation](https://pytorch.org/docs/stable/generated/torch.nn.utils.rnn.pack_sequence.html#torch.nn.utils.rnn.pack_sequence)


```python
class Model(nn.Module):
    def __init__(self):
        super().__init__()

        # embed_dim = 128
        embed_dim = 64
        
        d = 128
        # d = 512

        classes_n = 256

        self.embedding = nn.Embedding(classes_n + 1, embed_dim, padding_idx=0)
        #: classes_n+1: because padding adds another token

        self.rnn = nn.GRU(input_size=embed_dim, hidden_size=d, num_layers=2, batch_first=True, dropout=0.3)
        self.fc = nn.Linear(d, classes_n)

    def forward_packed(self, x):
        #: much much slower on CPU on my machine
        ##
        # Get sequence lengths
        seq_lengths = torch.tensor([len(seq) for seq in x], dtype=torch.long, device=x[0].device)
        # assert torch.all(seq_lengths > 0)
        
        # Pad sequences
        x_padded = torch.nn.utils.rnn.pad_sequence(x, batch_first=True, padding_value=0)
        
        # Embedding
        x_embedded = self.embedding(x_padded)
        
        # Pack sequences
        x_packed = torch.nn.utils.rnn.pack_padded_sequence(x_embedded, seq_lengths, batch_first=True, enforce_sorted=False)
        
        # RNN
        x_rnn, _ = self.rnn(x_packed)
        
        # Unpack sequences
        x_unpacked, _ = torch.nn.utils.rnn.pad_packed_sequence(x_rnn, batch_first=True)
        
        # Fully connected layer
        x_fc = self.fc(x_unpacked)
        
        return x_fc
    
    
    def forward(self, x, hidden=None):
        #: x: list of tensors shaped (seq_length, 256)
        ##
        x = torch.nn.utils.rnn.pad_sequence(x, batch_first=True, padding_value=0.0)
        #: returns B x T x * where T is the length of the longest sequence.
        # assert not has_nan(x)
        
        x = self.embedding(x) #: (B, T, embed_dim)
        # assert not has_nan(x)
        # ic('embedding: ', torch_shape_get(x))

        x, hidden = self.rnn(x, hidden) #: (B, T, d)
        # assert not has_nan(x)
        
        x = self.fc(x) #: (B, T, classes_n)
        # assert not has_nan(x)
        
        return x, hidden
```

```python
def loss_fn(y, y_hat):
    #: y: list of tensors shaped (seq_length, classes_n+1)
    #: y_hat: tensor shaped (B, T, classes_n)
    ##
    y = torch.nn.utils.rnn.pad_sequence(y, batch_first=True, padding_value=0.0)
    #: returns B x T where T is the length of the longest sequence.
    # ic(y.shape, y.dtype)

    y_all = y.view(-1)
    # ic(y_all.shape, y_all.dtype)

    mask = (y_all != 0).float()

    y_all_corrected = (y_all - 1) * mask
    #: subtracted one from y to make the targets start from zero

    y_hat_all = y_hat.view(-1, y_hat.size(-1))
    # ic(y_hat_all.shape, y_hat_all.dtype)
    
    criterion = nn.CrossEntropyLoss(reduction='none')
    loss = criterion(y_hat_all, y_all_corrected.to(torch.long))

    masked_loss = loss * mask

    average_loss = torch.sum(masked_loss) / torch.sum(mask)

    return average_loss
```

```python
def shift_left(tensor_list, pad_value=0.0):
    shifted_tensors = []
    for tensor in tensor_list:
        # Remove the first element from the input tensor
        shifted_tensor = tensor[1:]

        # Create a padding tensor with the same size as the last dimension of the input tensor
        padding = torch.full_like(tensor[-1:], pad_value)

        # Concatenate the shifted tensor and the padding tensor along the first dimension (rows)
        target_tensor = torch.cat((shifted_tensor, padding), dim=0)

        shifted_tensors.append(target_tensor)
    
    return shifted_tensors

# Example usage:
input_ids = [torch.tensor([1, 2, 3, 4]), torch.tensor([5, 6, 7, 8])]
print("Input Ids:")
print(input_ids)

target_ids = shift_left(input_ids)
print("Shifted Left (Target Ids):")
print(target_ids)
```

```python
import torch
import heapq

def tensor_to_string(tensor):
    chars = [chr(c) for c in tensor]
    return ''.join(chars)

def tensor_append_scalar(tensor, scalar):
    scalar_tensor = torch.tensor(scalar).view(1)  # Add a dimension to match the original tensor's dimensions
    scalar_tensor = scalar_tensor.to(device)

    # Append the scalar to the original tensor
    result = torch.cat((tensor, scalar_tensor), dim=0)
    return result


def generate_next_top_k(model, input_sequence, k):
    logits, _ = model.forward([input_sequence])
    logits = logits[0, -1, :]
    # ic(torch_shape_get(logits))
    
    probabilities = torch.softmax(logits, dim=-1)
    # ic(torch_shape_get(probabilities))

    top_k_values, top_k_indices = torch.topk(probabilities, k)

    return [(tensor_append_scalar(input_sequence, idx.item() + 1), log_prob.item()) for idx, log_prob in zip(top_k_indices, top_k_values.log())]

def beam_search(model, desired_length, starting_string, k=5):
    with ModelEvalMode(model), torch.no_grad():
      input_sequence = torch.tensor(str_to_np(starting_string).astype(np.int32) + 1, dtype=torch.long)
      input_sequence = input_sequence.to(device)
      # ic(torch_shape_get(input_sequence))
      
      log_prob = 0.0

      beam = [(input_sequence, log_prob)]

      while len(beam[0][0]) < desired_length:
          new_beam = []
          for seq, log_prob in beam:
              next_top_k = generate_next_top_k(model, seq, k)
              new_beam.extend([(new_seq, new_log_prob + log_prob) for new_seq, new_log_prob in next_top_k])

          beam = heapq.nlargest(k, new_beam, key=lambda x: x[1])

      return [tensor_to_string(seq - 1) for seq, _ in beam]
```

```python
def eval_gen(*args, display=999999, **kwargs):
    generated_texts = beam_search(
        *args, **kwargs,
    )

    for idx, text in enumerate(generated_texts):
        if idx >= display:
            break
        
        print(f"Generated text {idx + 1}: {text}")
```

```python
#: Training Loop

if torch.cuda.is_available():
	device = 'cuda'
	non_blocking = True
elif True:
	device = 'cpu'
	non_blocking = False
else:
	#: causes NaNs
	device = 'mps' 
	non_blocking = False

i = 0
epochs = 3000
batch_size = 4096
# batch_size = 3
learning_rate = 0.005
tbptt_steps = 64
max_len = 0

m = Model().to(device=device, non_blocking=non_blocking)
m.train()

optimizer = torch.optim.AdamW(m.parameters(), lr=learning_rate)

counter = 0
for epoch in range(epochs):
	dt = dt.shuffle()
	for i in range(0, len(dt), batch_size):
		batch = dt[i:i+batch_size]
		inputs = batch['input']

		# for input in inputs:
			# assert not has_nan(input)

		if max_len > 0:
			inputs = list(map(lambda x: x[:max_len] if len(x) > max_len else x, inputs))

		lens = [len(seq) for seq in inputs]
		current_max_len = max(lens)
		mean_len = statistics.mean(lens)
		# ic(current_max_len, mean_len)

		inputs = list(map(lambda x: x.to(device, non_blocking=non_blocking), inputs))
		# ic(torch_shape_get(inputs))

		targets = shift_left(inputs)
		# ic(torch_shape_get(targets))

		##
		# output_logits, _ = m.forward(inputs)
		
		# l = loss_fn(targets, output_logits)

		# optimizer.zero_grad()
		# l.backward()
		# torch.nn.utils.clip_grad_norm_(m.parameters(), max_norm=1)
		# optimizer.step()
		##
		hidden = None #: (num_layers, batch_size, hidden_size)
		optimizer.zero_grad()
		total_loss = 0
		for t in range(0, current_max_len, tbptt_steps):
			inputs_chunk = [seq[t:t+tbptt_steps] for seq in inputs]
			targets_chunk = [seq[t:t+tbptt_steps] for seq in targets]

			# Filter out empty sequences
			non_empty_indices = [i for i, chunk in enumerate(inputs_chunk) if len(chunk) > 0]
			if not non_empty_indices:
				continue
			inputs_chunk = [inputs_chunk[i] for i in non_empty_indices]
			targets_chunk = [targets_chunk[i] for i in non_empty_indices]

			hidden_current = hidden
			if hidden_current is not None:
				hidden_current = hidden_current[:, non_empty_indices]
				
			output_logits, hidden_next = m.forward(inputs_chunk, hidden_current)
			l = loss_fn(targets_chunk, output_logits)
			total_loss += l

			l.backward()
			torch.nn.utils.clip_grad_norm_(m.parameters(), max_norm=1)

			# Update the hidden state for non-empty sequences
			if hidden is not None:
				hidden[:, non_empty_indices] = hidden_next.detach()
			else:
				hidden = hidden_next.detach()
				#: The first time all indices should be nonempty.
				
		optimizer.step()
		
		# l = total_loss
		l = total_loss / mean_len
		##


		if counter % 1 == 0:
			l = l.item()
			print(f"loss: {l:>7f}  [{counter:>5d}, epoch={epoch}]")
		
		counter += 1

	# print(f"loss: {l:>7f}  [{counter-1:>5d}, epoch={epoch} finished!]")
	if epoch % 15 == 0:
		eval_gen(display=3, model=m, desired_length=100, starting_string="Harry ", k=32)
None
```
